```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("amazon.csv")
```

1. What is the average rating for each product category


```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
print("\n1. Average rating for each product category:")
print(df.groupby('category')['rating'].mean().reset_index())
```

2. What are the top rating_count products by category


```python
# Convert 'rating_count' to numeric, handling errors
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Group by 'category' and find the top product by 'rating_count'
print("\n2. Top rating_count products by category:")
print(df.groupby('category').apply(lambda x: x.nlargest(1, 'rating_count')).reset_index(drop=True))
```

3. What is the distribution of discounted prices vs. actual prices?

```python
# Assuming 'df' is your DataFrame loaded from 'amazon.csv'
# and 'discounted_price' and 'actual_price' are the column names

# Convert price columns to numeric (if needed)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Distribution of Discounted Prices
sns.histplot(df['discounted_price'], ax=axes[0], kde=True)
axes[0].set_title('Distribution of Discounted Prices')
axes[0].set_xlabel('Discounted Price')
axes[0].set_ylabel('Frequency')

# Distribution of Actual Prices
sns.histplot(df['actual_price'], ax=axes[1], kde=True)
axes[1].set_title('Distribution of Actual Prices')
axes[1].set_xlabel('Actual Price')
axes[1].set_ylabel('Frequency')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()

# Optional: Calculate and print summary statistics
print("Discounted Price Statistics:")
print(df['discounted_price'].describe())
print("\nActual Price Statistics:")
print(df['actual_price'].describe())
```

4. How does the average discount percentage vary across categories?



```python
# Assuming 'df' is your DataFrame loaded from 'amazon.csv'
# and 'discounted_price' and 'actual_price' are the column names

# Convert price columns to numeric (if needed)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Calculate discount percentage
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100

# Group by category and calculate average discount percentage
average_discount_by_category = df.groupby('category')['discount_percentage'].mean().reset_index()

# Print the results
print("Average Discount Percentage by Category:")
print(average_discount_by_category)

# Optional: Sort by average discount percentage
average_discount_by_category = average_discount_by_category.sort_values(by='discount_percentage', ascending=False)
print("\nAverage Discount Percentage by Category (Sorted):")
print(average_discount_by_category)


# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Distribution of Discounted Prices
sns.histplot(df['discounted_price'], ax=axes[0], kde=True)
axes[0].set_title('Distribution of Discounted Prices')
axes[0].set_xlabel('Discounted Price')
axes[0].set_ylabel('Frequency')

# Distribution of Actual Prices
sns.histplot(df['actual_price'], ax=axes[1], kde=True)
axes[1].set_title('Distribution of Actual Prices')
axes[1].set_xlabel('Actual Price')
axes[1].set_ylabel('Frequency')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()

# Optional: Calculate and print summary statistics
print("Discounted Price Statistics:")
print(df['discounted_price'].describe())
print("\nActual Price Statistics:")
print(df['actual_price'].describe())
```

6. What are the most popular product keywords?



```python
!pip install nltk
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Download the missing 'punkt_tab' resource


# Assuming 'df' is your DataFrame loaded from 'amazon.csv'
# and 'product_name' is the column name

# Function to extract keywords from a product name
def extract_keywords(product_name):
    tokens = word_tokenize(product_name.lower())
    stop_words = set(stopwords.words('english'))
    keywords = [token for token in tokens if token.isalnum() and token not in stop_words]
    return keywords

# Extract keywords for all product names
df['keywords'] = df['product_name'].apply(extract_keywords)

# Combine all keywords into a single list
all_keywords = [keyword for keywords_list in df['keywords'] for keyword in keywords_list]

# Calculate keyword frequency
keyword_frequency = Counter(all_keywords)

# Print the top N most popular keywords (e.g., top 10)
N = 10  # Change this to the desired number of top keywords
print(f"Top {N} Most Popular Keywords:")
for keyword, frequency in keyword_frequency.most_common(N):
    print(f"{keyword}: {frequency}")
```

6. Among the top 5 most popular artists, which artist has the highest popularity on average?
 Calculate and display the average popularity for each artist.


```python
# Get the top 5 most popular artists
top_5_artists = df['Artist'].value_counts().head(5).index

# Calculate the average popularity for each of the top 5 artists
average_popularity = df[df['Artist'].isin(top_5_artists)].groupby('Artist')['Popularity'].mean()

# Display the result
print(average_popularity)

# Find the artist with the highest average popularity
artist_with_highest_avg_popularity = average_popularity.idxmax()
print(f"\nThe artist with the highest average popularity among the top 5 is: {artist_with_highest_avg_popularity}")
```

7. For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.


```python
# Get the top 5 most popular artists
top_5_artists = df['Artist'].value_counts().head(5).index

# Find the most popular track for each artist
for artist in top_5_artists:
    most_popular_track = df[df['Artist'] == artist].sort_values('Popularity', ascending=False)['Track Name'].iloc[0]
    print(f"Most popular track for {artist}: {most_popular_track}")
```

10. How does the distribution of track popularity vary for different artists?
Visualize this using a swarm plot or a violin plot.

Answer: Swarm plots show each individual data point, allowing you to see the distribution and potential clusters of popularity for each artist. They can become cluttered with large datasets.


```python
# Create the swarm plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.swarmplot(x='Artist', y='Popularity', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.title('Track Popularity Distribution Across Artists (Swarm Plot)')
plt.show()
```

Violin plots provide a smoother representation of the density distribution of popularity for each artist, making it easier to compare the overall shapes of the distributions.

```python
# Create the violin plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.violinplot(x='Artist', y='Popularity', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.title('Track Popularity Distribution Across Artists (Violin Plot)')
plt.show()
```
